```python
# Class for Policyholders
class Policyholder:
    def __init__(self, name, age, address, policy_number):
        self.name = name
        self.age = age
        self.address = address
        self.policy_number = policy_number
        self.status = "Active"  # Default status is Active when created

    def register(self):
        self.status = "Active"
        print(f"Policyholder {self.name} (Policy Number: {self.policy_number}) has been registered and is now Active.")

    def suspend(self):
        if self.status != "Suspended":
            self.status = "Suspended"
            print(f"Policyholder {self.name} (Policy Number: {self.policy_number}) has been suspended.")
        else:
            print(f"Policyholder {self.name} is already suspended.")

    def reactivate(self):
        if self.status == "Suspended":
            self.status = "Active"
            print(f"Policyholder {self.name} (Policy Number: {self.policy_number}) has been reactivated.")
        else:
            print(f"Policyholder {self.name} is already Active.")

    def update_address(self, new_address):
        self.address = new_address
        print(f"Address updated for {self.name}.")

    def display_details(self):
        print(f"Policyholder Details:")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Address: {self.address}")
        print(f"Policy Number: {self.policy_number}")
        print(f"Status: {self.status}")


# Class for Products
class Product:
    all_products = {}  # Class-level dictionary to manage all products

    def __init__(self, product_id, name, coverage_amount, premium):
        self.product_id = product_id
        self.name = name
        self.coverage_amount = coverage_amount
        self.premium = premium
        self.status = "Active"  # Default status is Active when created

    @classmethod
    def create_product(cls, product_id, name, coverage_amount, premium):
        if product_id in cls.all_products:
            print(f"Product with ID {product_id} already exists.")
        else:
            product = Product(product_id, name, coverage_amount, premium)
            cls.all_products[product_id] = product
            print(f"Product {name} (ID: {product_id}) created successfully.")

    @classmethod
    def update_product(cls, product_id, name=None, coverage_amount=None, premium=None):
        product = cls.all_products.get(product_id)
        if product:
            if name:
                product.name = name
            if coverage_amount:
                product.coverage_amount = coverage_amount
            if premium:
                product.premium = premium
            print(f"Product {product_id} updated successfully.")
        else:
            print(f"Product with ID {product_id} not found.")

    @classmethod
    def remove_product(cls, product_id):
        product = cls.all_products.get(product_id)
        if product:
            product.status = "Suspended"
            print(f"Product {product_id} has been suspended.")
        else:
            print(f"Product with ID {product_id} not found.")

    @classmethod
    def display_all_products(cls):
        if cls.all_products:
            print("All Products:")
            for product in cls.all_products.values():
                product.display_product()
        else:
            print("No products available.")

    def display_product(self):
        print(f"Product Details:")
        print(f"Product ID: {self.product_id}")
        print(f"Name: {self.name}")
        print(f"Coverage Amount: {self.coverage_amount}")
        print(f"Premium: {self.premium}")
        print(f"Status: {self.status}")


# Class for Payments
class Payment:
    def __init__(self, payment_id, policy_number, amount, payment_date, due_date, penalty_rate=0.05):
        self.payment_id = payment_id
        self.policy_number = policy_number
        self.amount = amount
        self.payment_date = payment_date
        self.due_date = due_date
        self.penalty_rate = penalty_rate
        self.is_paid = False

    def process_payment(self):
        if not self.is_paid:
            self.is_paid = True
            print(f"Payment of {self.amount} for policy {self.policy_number} processed on {self.payment_date}.")
        else:
            print(f"Payment for policy {self.policy_number} has already been processed.")

    def send_reminder(self, current_date):
        if not self.is_paid:
            print(f"Reminder: Payment of {self.amount} for policy {self.policy_number} is due by {self.due_date}. Current date: {current_date}.")
        else:
            print(f"No reminder needed. Payment has already been made for policy {self.policy_number}.")

    def calculate_penalty(self, current_date):
        if not self.is_paid and current_date > self.due_date:
            overdue_days = (current_date - self.due_date).days
            penalty = self.amount * self.penalty_rate * overdue_days
            print(f"Payment is overdue by {overdue_days} days. Penalty accrued: {penalty:.2f}.")
            return penalty
        elif self.is_paid:
            print(f"No penalty. Payment has been made.")
            return 0
        else:
            print(f"No penalty yet. Payment is not overdue.")
            return 0

    def display_payment(self):
        print(f"Payment Details:")
        print(f"Payment ID: {self.payment_id}")
        print(f"Policy Number: {self.policy_number}")
        print(f"Amount: {self.amount}")
        print(f"Payment Date: {self.payment_date}")
        print(f"Due Date: {self.due_date}")
        print(f"Status: {'Paid' if self.is_paid else 'Unpaid'}")
```

```python
# Class for Policyholders
class Policyholder:
    def __init__(self, name, age, address, policy_number):
        self.name = name
        self.age = age
        self.address = address
        self.policy_number = policy_number
        self.status = "Active"  # Default status is Active when created
        self.payments = []  # List to track payments

    def register(self):
        self.status = "Active"
        print(f"Policyholder {self.name} (Policy Number: {self.policy_number}) has been registered and is now Active.")

    def suspend(self):
        if self.status != "Suspended":
            self.status = "Suspended"
            print(f"Policyholder {self.name} (Policy Number: {self.policy_number}) has been suspended.")
        else:
            print(f"Policyholder {self.name} is already suspended.")

    def reactivate(self):
        if self.status == "Suspended":
            self.status = "Active"
            print(f"Policyholder {self.name} (Policy Number: {self.policy_number}) has been reactivated.")
        else:
            print(f"Policyholder {self.name} is already Active.")

    def add_payment(self, payment):
        self.payments.append(payment)

    def display_details(self):
        print(f"\nPolicyholder Details:")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Address: {self.address}")
        print(f"Policy Number: {self.policy_number}")
        print(f"Status: {self.status}")
        if self.payments:
            print(f"\nPayments:")
            for payment in self.payments:
                payment.display_payment()
        else:
            print("No payments found.")


# Class for Products
class Product:
    all_products = {}

    def __init__(self, product_id, name, coverage_amount, premium):
        self.product_id = product_id
        self.name = name
        self.coverage_amount = coverage_amount
        self.premium = premium
        self.status = "Active"

    @classmethod
    def create_product(cls, product_id, name, coverage_amount, premium):
        if product_id in cls.all_products:
            print(f"Product with ID {product_id} already exists.")
        else:
            product = Product(product_id, name, coverage_amount, premium)
            cls.all_products[product_id] = product
            print(f"Product {name} (ID: {product_id}) created successfully.")

    @classmethod
    def display_all_products(cls):
        if cls.all_products:
            print("\nAll Products:")
            for product in cls.all_products.values():
                product.display_product()
        else:
            print("No products available.")

    def display_product(self):
        print(f"\nProduct Details:")
        print(f"Product ID: {self.product_id}")
        print(f"Name: {self.name}")
        print(f"Coverage Amount: {self.coverage_amount}")
        print(f"Premium: {self.premium}")
        print(f"Status: {self.status}")


# Class for Payments
class Payment:
    def __init__(self, payment_id, policy_number, amount, payment_date, due_date):
        self.payment_id = payment_id
        self.policy_number = policy_number
        self.amount = amount
        self.payment_date = payment_date
        self.due_date = due_date
        self.is_paid = True

    def display_payment(self):
        print(f"\nPayment Details:")
        print(f"Payment ID: {self.payment_id}")
        print(f"Policy Number: {self.policy_number}")
        print(f"Amount: {self.amount}")
        print(f"Payment Date: {self.payment_date}")
        print(f"Due Date: {self.due_date}")
        print(f"Status: {'Paid' if self.is_paid else 'Unpaid'}")


# Create Products
Product.create_product("PROD0001", "Car Insurance", 1000000, 50000)
Product.create_product("PROD0002", "Health Insurance", 20000000, 700000)

# Create Policyholders
policyholder1 = Policyholder("Paul Kimathi", 30, "Plot 11-12 George Street", "POL012589")
policyholder2 = Policyholder("Jayden Lorenzo", 35, "Plot 37 Kings Avenue", "POL012590")

# Create Payments for the Policyholders
from datetime import datetime

payment1 = Payment(
    payment_id="PAY001",
    policy_number="POL012589",
    amount=50000,
    payment_date=datetime(2025, 1, 10),
    due_date=datetime(2025, 1, 31)
)

payment2 = Payment(
    payment_id="PAY002",
    policy_number="POL012590",
    amount=700000,
    payment_date=datetime(2025, 1, 12),
    due_date=datetime(2025, 1, 31)
)

# Add Payments to Policyholders
policyholder1.add_payment(payment1)
policyholder2.add_payment(payment2)

# Display Details
Product.display_all_products()
policyholder1.display_details()
policyholder2.display_details()
```
